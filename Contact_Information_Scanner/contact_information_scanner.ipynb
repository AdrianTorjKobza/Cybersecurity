```python
import os, re
from zipfile import ZipFile
```

```python
# Define regex rules for email and phone number.

email_regex = r'[a-z0-9]+[\._]?[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}'
phone_regex = r'^\+\d{1,4}(?: ?\d){10}$'

regexes = [email_regex, phone_regex]
```

```python
# Check for emails and phone numbers within the files.

def find_contact_information(data):
  matches = []

  for regex in regexes:
    m = re.findall(regex, data)
    matches += m

  return matches
```

```python
def print_matches(filedir, matches):
  if len(matches) > 0:
    print(filedir)

    for match in matches:
      print(match)

  else:
    print("No matches found.")
```

```python
# Parse the files.

def parse_text(root, txts):
  for txt in txts:
    filedir = os.path.join(root, txt)

    with open(filedir, 'r') as f:
      data = f.read()

    matches = find_contact_information(data)
    print_matches(filedir, matches)
```

```python
file_ext = [".txt", ".cvs"] # File extensions to search for.
```

```python
# Search for files with the .txt and .cvs extension and start parsing them.

def find_files(directory):
  for root, dirs, files in os.walk(directory):
    for ext in file_ext:
      parse_text(root, [f for f in files if f.endswith(ext)])
```

```python
directory = os.path.join(os.getcwd(), "Documents")

find_files(directory)
```
