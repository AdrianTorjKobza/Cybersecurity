```python
from scapy.all import * # Scapy can build, send, and receive raw network packets.
```

```python
ports = [25, 80, 53, 443, 445, 8080, 8443] # The TCP ports to be scanned.
```

```python
# Probe the TCP ports on host using SYN packets.

def syn_scan(host):
  # Send packets and receive answers.
  ans, unans = sr(IP(dst=host)/TCP(sport=5555, dport=ports, flags="S"),
                  timeout=2,
                  verbose=1)

  print("Open ports at %s:" % host)

  # Iterate over the sent (s) and received (r) pair packages.
  for (s, r) in ans:
    if s[TCP].dport == r[TCP].sport:
      print ("Port: " + str(s[TCP].dport))
```

```python
# Test whether a DNS server at host will answer a DNS query.

def dns_scan(host):
  ans, unans = sr(IP(dst=host)/UDP(sport=5555, dport=53)/DNS(rd=1, qd=DNSQR (qname="google.com")),
                  timeout = 2,
                  verbose=1)

  if ans:
    print("\nDNS Server at %s"%host)
```

```python
# Target host IP (Google Public DNS).
host = "8.8.8.8"
```

```python
syn_scan(host)
dns_scan(host)
```
